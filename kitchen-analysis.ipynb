```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
path = r"C:\Users\Rudra\Desktop\cloud-kitchen-dashboard\kittchen-pnl-data.xlsx"
df = pd.read_excel(path, header=1)
```

```python
df.columns = (df.columns.str.strip().str.upper().str.replace(" ", "_"))   
```

```python
df[['MONTH_NAME', 'YEAR']]= df['MONTH'].str.split("-", expand=True)
```

```python
df.head(2)
```

```python
df['GM'] = df['GROSS_MARGIN']
df['GM_%'] = ( df['GM'] / df['NET_REVENUE'] ) * 100

df["CM"] = df["NET_REVENUE"] - df["IDEAL_FOOD_COST"] - df["DISCOUNT"]
df['CM_%'] = ( df['CM'] / df['NET_REVENUE'] ) * 100

df['EBITDA'] = df['KITCHEN_EBITDA']
```

```python
sns.barplot(
    df,
    x=df['CITY'],
    y=df['NET_REVENUE']
)

plt.show()
```

```python
sns.barplot(
    df,
    x=df['ZONE_MAPPING'],
    y=df['NET_REVENUE']
)

plt.show()
```

```python
def barplot_with(col):
    sns.barplot(
        df,
        x=df[col],
        y=df['NET_REVENUE']
    )

    plt.show()
```

```python
df.info()
```

```python
for col in ['REVENUE_COHORT','CM_COHORT', 'EBITDA_CATEGORY', 'EBITDA_COHORT', 'MONTH_NAME', 'YEAR']:
    barplot_with(col)
```

```python
pd.pivot_table(
    df,
    values= [ "NET_REVENUE", "GM", "CM", "EBITDA" ],
    index= ["REVENUE_COHORT", "CM_COHORT", "EBITDA_CATEGORY", "EBITDA_COHORT"],
    columns=["MONTH_NAME", "YEAR"],
    aggfunc="sum",
    fill_value=0
)
```

```python
df.columns
```

```python
df['VARIANCE'].describe()
```

```python
bins = [10019, 15495, 20195, 25181, 29994]

labels = ['Low', 'Medium-Low', 'Medium-High', 'High']

# Apply pd.cut to your column
df['VARIANCE_CATEGORY'] = pd.cut(df['VARIANCE'], bins=bins, labels=labels, include_lowest=True)
```

```python
variance_pivot_table =  pd.pivot_table(
    df,
    values="VARIANCE",
    index="REVENUE_COHORT",
    columns=["YEAR", "MONTH_NAME"],
    aggfunc="sum",
    fill_value=0
)

variance_pivot_table
```

```python
pd.DataFrame(
    variance_pivot_table.sum(axis=0)
).T
```

```python
# store count

store_count_table =  pd.pivot_table(
    df,
    values="STORE",
    index="VARIANCE_CATEGORY",
    columns=["YEAR", "MONTH_NAME"],
    aggfunc="count",
    fill_value=0,
    observed=True
)

store_count_table
```

```python
pd.DataFrame(
    store_count_table.sum(axis=0)
).T
```
